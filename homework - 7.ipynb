```python
import json
y = {}
with open('purchase_log.txt', 'r', encoding = 'utf-8') as f:
    for i,line in enumerate( f):
        if i != 0:
            y[json.loads(line)["user_id"]] = json.loads(line)["category"]
with open('visit_log.csv', 'r', encoding = 'utf-8') as f3:
    with open('funnel.csv', 'w', encoding = 'utf-8') as f2write:
        for i,line in enumerate (f3):
            if i != 0: 
                if y.get(line.strip().split(',')[0]) != None:
                    f2write.write(line.strip().split(',')[0])
                    f2write.write(':')
                    f2write.write(y[line.strip().split(',')[0]])
                    f2write.write('\n')
                
```
